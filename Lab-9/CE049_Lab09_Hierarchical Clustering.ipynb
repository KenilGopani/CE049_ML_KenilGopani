```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("darkgrid")
```

```python
X, y = make_blobs(n_samples=100, centers=3, cluster_std = 1.00)
sns.scatterplot(x=X[:,0], y=X[:,1], c =["green"])
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
dendrogram(linkage(X, 'single'),
orientation='top',
distance_sort='descending',
show_leaf_counts=True)

plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc_model = AgglomerativeClustering(n_clusters=3, affinity='euclidean',linkage='ward')
hc_model.fit_predict(X)
```

```python
#pring the data points
plt.scatter(x=X[:,0], y=X[:,1], c= hc_model.labels_, cmap='rainbow' )
```

# Task on Diabetes Dataset

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA # PCA library

data = load_diabetes()
print("Features: ", data.feature_names)
```

```python
features = data.data
label = data.target
label
```

```python
pca = PCA(n_components=2)
pcaTr = pca.fit(features)
dataPCA = pcaTr.transform(features)
```

```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
dendrogram(linkage(dataPCA, 'single'),orientation='top',
distance_sort='descending',show_leaf_counts=True)
plt.show()
```

```python
# Calculate silhouette_score
from sklearn.metrics import silhouette_score
ac2 = AgglomerativeClustering(n_clusters = 2,affinity='euclidean',linkage='ward')
plt.figure(figsize =(6, 6))
plt.scatter(dataPCA[:,0], dataPCA[:,1], c = ac2.fit_predict(dataPCA), cmap ='rainbow')

plt.show()
print(silhouette_score(dataPCA, ac2.fit_predict(dataPCA)))
# print(ac2.fit_predict(dataPCA))
```
