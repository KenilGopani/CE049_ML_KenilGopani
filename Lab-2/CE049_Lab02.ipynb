### 1. Steps for Handling the missing value

1.Import Libraries
2.Load data
3.Seprate Input and Output attributes
4.Find the missing values and handle it in either way a. Removing data b. Imputation

```python
# Step 1: Import Libraries
import numpy as np 
import pandas as pd
from sklearn.impute import SimpleImputer 


# Step 2: Load Data
datasets = pd.read_csv('Exercise-CarData.csv') 
print("\nData :\n",datasets)
print("\nData statistics\n",datasets.describe())
```

```python
# Step 3: Seprate Input and Output attributes

#input attributes
# All rows, all columns except first
X = datasets.iloc[:,2:]

#output attributes
# All rows, only first column
Y = datasets.iloc[:,1:2]

print("\n\nInput : \n", X) 
print("\n\nOutput: \n", Y) 
```

```python
#axis value : 0 or 1 , 0 for row / 1 for column
datasets = (datasets.replace('??',np.nan))  #replace ?? with NaN 
datasets = (datasets.replace('????',np.nan))  #replace ???? with NaN 
#print(datasets)

# Step 4: Find the missing values and handle it in either way
# 4a. Removing the row with all null values in it's all columns
datasets.dropna(axis=0,how='all',inplace=True)
print("\nNew Data-1 :\n",datasets)


#4b. Removing the row with any one null values in it's columns
updated = datasets.dropna(axis=0,how='any',inplace=False)
print("\nNew Updated Data :\n",updated.count())
```

```python
print(datasets)
print(datasets.dtypes)
```

```python
#convert data type of KM & HP from object to float so that we can do mathematical task.
datasets['KM'] = datasets['KM'].apply(pd.to_numeric)
datasets['HP'] = datasets['HP'].apply(pd.to_numeric)
print(datasets.dtypes)
```

```python
#Now update NaN rows with values
updated_df = datasets;
# print(datasets.count())

updated_df['Age']=updated_df['Age'].fillna(updated_df['Age'].mean())
updated_df['KM']=updated_df['KM'].fillna(updated_df['KM'].mean())
updated_df['HP']=updated_df['HP'].fillna(updated_df['HP'].mean())
updated_df['MetColor']=updated_df['MetColor'].fillna(1.0)
updated_df['FuelType'] = updated_df['FuelType'].fillna("Petrol")

doors = updated_df.iloc[:,9:10]
doors[doors['Doors'] == "three"] = 3
doors[doors['Doors'] == "four"] = 4
doors[doors['Doors'] == "five"] = 5
updated_df['Doors'] = doors

print("Now total numbers of row in each column:-\n",datasets.count())
print("\n")
updated_df.info()
```

```python
#datasets after handling all missing values
print(updated_df.to_string())
# print(updated_df['MetColor'].where(updated_df['MetColor']==1.0).count())
```

#### Correlation is :
Statistical term which in common usage refers to how close two variables are to having a linear relationship with each other.

For example, two variables which are linearly dependent (say, **x** and **y** which depend on each other as x = 2y) will have a higher correlation than two variables which are non-linearly dependent (say, u and v which depend on each other as u = v2)
