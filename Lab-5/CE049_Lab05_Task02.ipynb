### Task 2: 
### Apply algorithm on digits dataset - One Hot Encoding of features: and Train test Division 65%-35%

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
```

```python
# load dataset
from sklearn.datasets import load_digits
digits = load_digits(as_frame=True)
```

```python
digits.data.shape
```

```python
digits.frame
```

### One-Hot Encoding

```python
# one-hot encoder
ohe = preprocessing.OneHotEncoder()

encoded_dataset = ohe.fit_transform(digits.frame).toarray()

print('shape of encoded dataset :')
print(encoded_dataset.shape)
print()
print('Encoded Dataset :')
print(encoded_dataset)
```

```python
data = pd.DataFrame(encoded_dataset)
data
```

### Spliting dataset

```python
from sklearn.model_selection import train_test_split

X = data.values[:,:-10]
Y = data.values[:,-10:]

# Train test Division 80%-20%
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.35, random_state = 49)
```

### Train ’Create and Train DecisionTreeClassifier’

```python
classifier = DecisionTreeClassifier(criterion = "gini", random_state = 49, max_depth=49, min_samples_leaf=5)

classifier.fit(X_train, y_train)
```

```python
# make prediction
y_pred = classifier.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn import metrics

# Accuracy
accuracy = metrics.accuracy_score(y_test,y_pred)
print('Accuracy : ',accuracy)

# Precision
precision = metrics.precision_score(y_test,y_pred, average=None)
print('Precision : ',precision)

# Recall
recall = metrics.recall_score(y_test,y_pred, average=None)
print('Recall : ',recall)
```
