```python
# Declare a figure with a custom size
fig = plt.figure(figsize=(7, 7))

mystrey_text = brown.words(categories='mystery')
fdist = nltk.FreqDist(w.lower() for w in mystrey_text)

# labels for the two classes
words = ['could', 'my', 'have','would','you','she','look','i']
data = []
for m in words:
    print(m + ':', fdist[m],",", end=' ')
    data.append(fdist[m])
    
# Declare pie chart, where the slices will be ordered and plotted counter-clockwise:
plt.pie(data, labels=words, autopct='%1.1f%%',shadow=True, startangle=90)
plt.show()
```

# Preprocess raw text for Sentiment analysis

```python
# Our selected sample
mystrey_text = brown.words(categories='mystery')
mytext = ""
for i in range(0,321):
    mytext = mytext+ mystrey_text[random.randint(0,10000)] + " "
print(mytext)
```

```python
#Let's import a few more libraries for this purpose
# download the stopwords from NLTK

nltk.download('stopwords')
```

```python
import re                                      # library for regular expression operations
import string                                  # for string operations
from nltk.corpus import stopwords              # module for stop words that come with NLTK
from nltk.stem import PorterStemmer            # module for stemming
from nltk.tokenize import TweetTokenizer       # module for tokenizing strings
```

```python
# remove hashtags
# only removing the hash # sign from the word
#print(mytext)
mystrey_text2 = re.sub(r'fifteen', '', mytext)
```

**Tokenize the string**
To tokenize means to split the strings into individual words without blanks or tabs. In this same step, we will also
convert each word in the string to lower case. The tokenize module from NLTK allows us to do these easily:

```python
# instantiate tokenizer class
tokenizer = TweetTokenizer(preserve_case=False)
# tokenize tweets
mytext_tokens = tokenizer.tokenize(mytext)
print()
print('Tokenized string:')
print(mytext_tokens)
```

**Remove stop words and punctuations**
The next step is to remove stop words and punctuation. Stop words are words that don't add significant
meaning to the text.

```python
#Import the english stop words list from NLTK
stopwords_english = stopwords.words('english')
print('Stop words\n')
print(stopwords_english)
print('\nPunctuation\n')
print(string.punctuation)
```

```python
print('\033[92m')
print("mytext_tokens:- ")
print(mytext_tokens)
print('\033[94m')
mytext_clean = []

for word in mytext_tokens:                       # Go through every word in your tokens list
    if (word not in stopwords_english and         # remove stopwords
        word not in string.punctuation):          # remove punctuation
        mytext_clean.append(word)
print('removed stop words and punctuation:')
print(mytext_clean)
```

**Stemming**
Stemming is the process of converting a word to its most general form, or stem. This helps in reducing the size
of our vocabulary.
NLTK has different modules for stemming and i will be using the PorterStemmer

```python
print()
print('\033[92m')
print(mytext_clean)
print('\033[94m')

stemmer = PorterStemmer()       # Instantiate stemming class    
mytext_stem = []               # Create an empty list to store the stems               
for word in mytext_clean:
    stem_word = stemmer.stem(word) # stemming word
    mytext_stem.append(stem_word) # append to the list
print('stemmed words:')
print(mytext_stem)
```
